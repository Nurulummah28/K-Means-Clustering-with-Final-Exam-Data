```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datanilai.csv", sep = ";")
df
```

```python
df.describe()
```

## Data training

```python
x_train = df[['UTS', 'UAS']].values
```

## Feature Scalling

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_train
```

## Clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3, random_state = 0)
kmeans
```

```python
y_cluster = kmeans.fit_predict(x_train)
y_cluster
```

```python
y_cluster
df['cluster']=y_cluster
```

```python
kmeans.cluster_centers_
```

## Visualisasi data hasil clustering

```python
plt.figure(figsize=(10,5))
plt.scatter(x_train[:,0], x_train[:,1], c =kmeans.labels_)
```

```python
df
```

## Mengganti label cluster menjadi kategori

```python
df["cluster"]= df["cluster"].astype("object")

df.loc[df.cluster==0, "cluster"] = "Baik"
df.loc[df.cluster==1, "cluster"] = "Buruk"
df.loc[df.cluster==2, "cluster"] = "Sangat baik"
df
```

```python
#cara lain ganti cluster
df["cluster"]=df["cluster"].replace(['Baik', 'Buruk', 'Sangat baik'], [0, 1, 2])
df
```

## Metode Inersia = mencari jumlah k terbaik

```python
##The elbow method itu nama metodenya karena patah, patah pertama adalah k terbaik
from sklearn.cluster import KMeans
cs = []
for i in range (1,10):
    kmeans = KMeans(n_clusters=i, init = 'k-means++',max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x_train)
    cs.append(kmeans.inertia_)
plt.plot(range(1,10), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```
